```python
# Using the chipotle dataset from the previous exercise, 
# create a bar chart that shows the 4 most popular items and the revenue produced by each.
```

```python
from env import host, user, password

def get_db_url(db, user=user, host=host, password=password):
    return f'mysql+pymysql://{user}:{password}@{host}/{db}'
```

```python
chipotle_query = '''
SELECT *
FROM orders
;'''

orders = pd.read_sql(chipotle_query, get_db_url('chipotle'))
orders.columns
```

```python
orders['item_price'] = orders.item_price.str.replace('$', '').astype(float)
```

```python
orders.head()
```

```python
orders['total price'] = orders.quantity * orders.item_price
```

```python
orders['total price'].sort_values()
```

```python
most_revenue = pd.DataFrame(orders.groupby('item_name')\
                            .item_price.agg('sum').sort_values(ascending=False).head(5))
most_revenue
most_revenue.columns
```

```python
top_4 = pd.DataFrame(orders.groupby('item_name')\
        .quantity.agg('sum').sort_values(ascending=False)\
        .head(4))
top_4
```

```python
top_4_revenue = pd.merge(top_4, most_revenue, on='item_name', how='outer')
top_4_revenue['revenue'] = top_4_revenue.quantity * top_4_revenue.item_price
top_4_revenue
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(data=top_4_revenue, x='item_name', y='revenue')
plt.title('Top 4 Items at Chipotle')
```
