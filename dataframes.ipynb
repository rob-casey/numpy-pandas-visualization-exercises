```python
from pydataset import data
import pandas as pd
import numpy as np
```

```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
#  Create a column named passing_english that indicates whether each student 
#  has a passing grade in english.

df.english >=70
df['passing_english'] = df.english >= 70
df.head()
```

```python
# Sort the english grades by the passing_english column. 
# How are duplicates handled? 

df.assign(passing_english = df.english >=70).sort_values(by='passing_english')
#duplicates are sorted/handled by index
```

```python
# Sort the english grades first by passing_english and then by student name. 
# All the students that are failing english should be first, and within the 
# students that are failing english they should be ordered alphabetically. 
# The same should be true for the students passing english. 
# (Hint: you can pass a list to the .sort_values method)

df.assign(passing_english = df.english >=70).sort_values(by=['passing_english','name']).head()
```

```python
# Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.

df.assign(overall_grade=(df.mean(axis=1)))
```

```python
# # Load the mpg dataset. 
# # Read the documentation for the dataset and use it for the following questions:

mpg = data('mpg')

# - How many rows and columns are there?

mpg.shape
```

```python
# - What are the data types of each column?

mpg.dtypes
```

```python
# - Summarize the dataframe with .info and .describe

mpg.info()
mpg.describe()
```

```python
# - Rename the cty column to city.

mpg.rename(columns={'cty':'city'}, inplace=True)
mpg
```

```python
# - Rename the hwy column to highway.

mpg.rename(columns={'hwy':'highway'}, inplace=True)
mpg
```

```python
# - Do any cars have better city mileage than highway mileage?

mpg[mpg.city > mpg.highway]

# there are no cars that have better city mileage than highway
```

```python
# Create a column named mileage_difference this column should 
# contain the difference between highway and city mileage for each car.

mpg = mpg.assign(mileage_difference =  mpg.highway - mpg.city)
mpg
```

```python
# - Which car (or cars) has the highest mileage difference?

mpg.mileage_difference.nlargest(n=2, keep='all')
```

```python
mpg[['year','manufacturer', 'model', 'mileage_difference']].sort_values(by='mileage_difference', ascending=False).head(2)
```

```python
# - Which compact class car has the lowest highway mileage? The best?

compact_cars = mpg[mpg['class'] == 'compact']
compact_cars[compact_cars.highway == compact_cars.highway.min()]
```

```python
# - Create a column named average_mileage that is the mean of the city and highway mileage.

average_mpg = mpg[['city','highway']].mean(axis=1)
mpg = mpg.assign(average_mileage = average_mpg)
mpg
```

```python
# - Which dodge car has the best average mileage? The worst?

dodge = mpg[mpg.manufacturer == 'dodge']
best_avg_mileage = dodge[dodge.average_mileage == dodge.average_mileage.max()]
best_avg_mileage
```

```python
# Load the Mammals dataset. 
# Read the documentation for it, and use the data to answer these questions:

Mammals = data('Mammals')
Mammals.head()
```

```python
# - How many rows and columns are there?

Mammals.shape
```

```python
# - Summarize the dataframe with .info and .describe

Mammals.info()
Mammals.describe()
```

```python
# - What is the the weight of the fastest animal?

Mammals.weight[Mammals.speed == Mammals.speed.max()]
```

```python
# - What is the overall percentage of specials?

round((Mammals.specials.sum() / len(Mammals) * 100),2)
```

```python
# - How many animals are hoppers that are above the median speed? 
#   What percentage is this?

median_speed = Mammals.median()['speed']
hoppers = Mammals[Mammals['hoppers'] == True]
above_med_spd = len(hoppers[hoppers.speed > median_speed])
above_med_spd
```

```python
round((ab_med_spd/ len(Mammals) * 100),2)
```
