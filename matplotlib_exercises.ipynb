```python
from matplotlib import pyplot as plt
import math
```

```python
# Exercise 1
Use matplotlib to plot the following equation:
y = x^2 − x + 2
You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, the origin.
```

```python
x = [-4, -3, -2, -1, 0, 1, 2, 3, 4]
y = [n ** 2 - n + 2 for n in x]

plt.plot(x, y)
plt.xlim(-10, 10)
plt.ylim(0, 15)
plt.annotate('Origin', 
             xy=(0,0), 
             xytext=(1,1), 
             arrowprops={'arrowstyle': '->'}
            )

plt.show()
```

```python
# Exercise 2

Create and label 4 separate charts for the following equations 
(choose a range for x that makes sense):

    y = √ x
    y = x ^ 3
    y = 2 ^ x
    y = 1 / (x + 1)
    
```

```python
# Variables

x = [0, 1, 2, 3]
y = x
y1 = [math.sqrt(n) for n in x]
y2 = [n ** 3 for n in x]
y3 = [2 ** n for n in x]
# y4 = [1 (for n in x + 1)]
```

```python
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y - x$')
plt.ylim(0,3)

plt.show()
```

```python
plt.plot(x, y1)
plt.xlabel('$x$')
plt.ylabel('$y = √x$')
plt.title('y = math.sqrt(n) for n in x')

plt.show()
```

```python
plt.plot(x, y2)
plt.xlabel('$x$')
plt.ylabel('$y = x^3$')
plt.title('y = n ** 3 for n in x')

plt.show()
```

```python
plt.plot(x, y3)
plt.xlabel('$x$')
plt.ylabel('$y = 2^x$')
plt.title('y = 2 ** n for n in x')
```
