```python
from matplotlib import pyplot as plt
import math
```

```python
# Exercise 1
Use matplotlib to plot the following equation:
y = x^2 âˆ’ x + 2
You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, the origin.
```

```python
x = [-4, -3, -2, -1, 0, 1, 2, 3, 4]
y = [n ** 2 - n + 2 for n in x]

plt.plot(x, y)
plt.xlim(-10, 10)
plt.ylim(0, 15)
plt.annotate('Origin', 
             xy=(0,0), 
             xytext=(1,1), 
             arrowprops={'arrowstyle': '->'}
            )

plt.show()
```

```python
# Exercise 2

Create and label 4 separate charts for the following equations 
(choose a range for x that makes sense):

    y = âˆš x
    y = x ^ 3
    y = 2 ^ x
    y = 1 / (x + 1)
```

```python
# Variables

x = [0, 1, 2, 3]
y = x
y1 = [math.sqrt(n) for n in x]
y2 = [n ** 3 for n in x]
y3 = [2 ** n for n in x]
y4 = [1/(n+1) for n in x]
```

```python
plt.plot(x, y1)
plt.xlabel('$x$')
plt.ylabel('$y = âˆšx$')
plt.title('y = math.sqrt(n) for n in x')

plt.show()
```

```python
plt.plot(x, y2)
plt.xlabel('$x$')
plt.ylabel('$y = x^3$')
plt.title('y = n ** 3 for n in x')

plt.show()
```

```python
plt.plot(x, y3)
plt.xlabel('$x$')
plt.ylabel('$y = 2^x$')
plt.title('y = 2 ** n for n in x')
```

```python
plt.plot(x, y4)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y = 1/(x+1)$')
```

```python
# Exercise 3
Combine the figures you created in the last step into one large figure with 4 subplots
```

```python
plt.figure(figsize=(16,8))

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('$y = âˆšx$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('$y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('$y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

```python
# Exercise 4
Combine the figures you created in the last step into one figure where 
each of the 4 equations has a different color for the points. 
Be sure to include a legend and an appropriate title for the figure.
```

```python
plt.figure(figsize=(16,8))

plt.plot(x, y1, c='red', label=('$y = âˆšx$'))

plt.plot(x, y2, c='blue', label=('$y = x^3$'))

plt.plot(x, y3, c='gold', label=('$y = 2^x$'))

plt.plot(x, y4, c='green', label=('$y = 1/(x+1)$'))

plt.title('Power Ranger Plots')
plt.legend()
plt.show()
```
