```python
from matplotlib import pyplot as plt
import math
```

```python
plt.figure(figsize=(16,8))

plt.xlabel('Elements')
plt.ylabel('Operations')
plt.title('Big O Notation')

x = range(1, 6)
y1 = [0*n + 1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n*math.log(n) for n in x]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]

plt.plot(x,y1, label='$\mathcal{O}(1)$')
plt.plot(x,y2, label='$\mathcal{O}(log n)$')
plt.plot(x,y3, label='$\mathcal{O}(n)$')
plt.plot(x,y4, label='$\mathcal{O}(n log n)$')
plt.plot(x,y5, label='$\mathcal{O}(n^2)$')
plt.plot(x,y6, label='$\mathcal{O}(2^n)$')
plt.plot(x,y7, label='$\mathcal{O}(n!)$')
plt.plot(x,y8, label='$\mathcal{O}(n^n)$')

plt.title('Big $\mathcal{O}$ Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.ylim(0, 20)
plt.xlim(1, 5)

# plt.legend()
plt.show()
```
